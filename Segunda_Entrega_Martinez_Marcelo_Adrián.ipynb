#### **ABSTRACTO**
El presente informe está destinado a los responsables de la gestión de la Clínica Tiempo la cual cuenta con una larga trayectoria en la ciudad, siendo la más prestigiosa de la región.

Para ello, el director de la clínica nos facilita este dataset con datos recopilados sobre distintas cuestiones de los pacientes que han estado en la clínica. 
Los campos del dataset se refieren a distintos aspectos registrados de cada paciente.
En base a ello y a las necesidades de información de los directivos, se realizará un modelo predictivo con el fin de establecer la duración de la estancia de los pacientes en la Clínica.

#### **OBJETIVOS**

#### **CONTEXTO ANALÍTICO**
Para cumplir con los objetivos planteados y dar solución a la problemática, el director de la clínica nos facilita este dataset con datos recopilados sobre distintas cuestiones de los pacientes que han estado en la clínica. El mismo está estructurado en 28 columnas y 100.000 filas. 
Los campos del dataset se refieren a distintos aspectos registrados de cada paciente, entre las variables están: 
*   ID Admisión: un identificador por cada admisión de un paciente que se da en el hospital.
*   FechaVisita: la fecha en la que se produjo la visita del paciente.
*   NumReadmisiones: el número de readmisiones que se han dado. 
*   Género: 1 para mujeres y 0 para hombres.
*   Renal: si tiene algún síntoma o problema de índole renal.
*   Asma: si tiene algún síntoma o problema de asma.
*   FaltaHierro: si tiene algún problema de falta de hierro o no.
*   Neumonia: si tiene algún problema de neumonía o no.
*   DependenciaSustancias: si tiene algún problema de dependencia de sustancias o no.
*   DesordenPsicologico: si tiene algún problema de desórdenes psicológicos o no.
*   Depresion: si tiene algún problema de depresión o no.
*   OtrosDesordenesPsiquicos: si tiene algún problema de otros desórdenes psíquicos o no.
*   Fibrosis: si tiene algún problema de fibrosis o no.
*   Malnutricion: si tiene algún problema de malnutrición o no.
*   DesordenSangre: si tiene algún problema de desorden en sangre o no.
*   Sodio: si tiene algún síntoma o problema de sodio.
*   Hematocrito: si tiene algún síntoma o problema de hematocrito.
*   Neutrofilos: si tiene algún síntoma o problema de neutrófilos.
*   Glucosa: si tiene algún síntoma o problema de glucosa.
*   UreaSangre: si tiene algún síntoma o problema de urea en sangre.
*   Creatinina: si tiene algún síntoma o problema de creatinina.
*   BMI: el índice de masa corporal del paciente.
*   Pulso: pulsaciones del paciente por minuto.
*   Respiracion: la respiración del paciente cada 30 segundos.
*   DiagnosticoSecundario: cuál es el diagnóstico secundario del paciente.
*   FechaAlta: la fecha en la que se le ha dado de alta al paciente.
*   IDHospital: el identificar de la sede del hospital donde ha sido ingresado.
*   DuracionEstancia: los días que duró la estancia del paciente en la clínica.

**Gráfico 1: Cantidad de veces que fueron readmitidos los pacientes**

```python
CANTIDAD_READMISIONES = Clinica.groupby(['numReadmisiones']).size().reset_index(name='Cantidad_Readmisiones')

fig, ax = plt.subplots(figsize=(8,4))
plt.xticks(range(18))
ax.bar(CANTIDAD_READMISIONES['numReadmisiones'], CANTIDAD_READMISIONES.Cantidad_Readmisiones)
ax.set_title('Gráfico 1: Cantidad de veces que fueron readmitidos los pacientes')
ax.set_ylabel('Cantidad de Pacientes')
ax.set_xlabel('Cantidad de Readmisiones')
```

**Interpretación**

En primera instancia, un dato interesante que se puede extraer del conjunto está vinculado con el número de readmisiones de los pacientes.
De ello se desprende que, el 57,92% de los pacientes no tuvieron readmisiones; en tanto que el 15,79% fue readmitido al menos una vez; el 10,51% fue reamitido dos veces; el 8,47% fue readmitido 3 veces y el restante 7,31% fue readmitido 4 veces.


##### **Gráfico 2: Porcentaje de pacientes según el género**
Un dato importante a señalar es la cantidad de pacientes que estuvieron internados en la clínica según su género.

```python
CANTIDAD_PACIENTES_GENERO = Clinica.groupby(['genero']).size().reset_index(name='Cantidad_por_Genero')

#x = [54750 , 40263]
labels = ['MASCULINO' , 'FEMENINO']
colors = ['tab:blue',  'tab:orange']

fig, ax = plt.subplots()
ax.pie(CANTIDAD_PACIENTES_GENERO['Cantidad_por_Genero'], labels = labels, 
       colors = colors, 
       autopct='%.0f%%')
ax.set_title('Gráfico 2: Porcentaje de pacientes según el género')
# círculo blanco en el centro
my_circle=plt.Circle( (0,0), 0.7, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

###### **Interpretación:**
Como se puede observar en el **Gráfico 1**, el **58%** (54.750) de los pacientes corresponden al **género femenino** y el restante **42%** (40.263) al **masculino**; esto nos indica que en mayor proporción los pacientes son mujeres.

### **Gráfico 3: Cantidad de pacientes según la duración de la estadía**
Otro aspecto relevante y que se vincula con uno de los objetivos que se plantean, se basa en cuál es la cantidad de pacientes agrupados en función de la cantidad de días que permanecieron en la clínica.

```python
CANTIDAD_PACIENTES_ESTADIA = Clinica.groupby(['duracionEstancia']).size().reset_index(name='Cantidad_Pacientes').head(10)
CANTIDAD_PACIENTES_ESTADIA
```

```python
fig, ax = plt.subplots(figsize=(8,4))
plt.xticks(range(18))
ax.bar(CANTIDAD_PACIENTES_ESTADIA['duracionEstancia'], CANTIDAD_PACIENTES_ESTADIA.Cantidad_Pacientes)
ax.set_title('Gráfico 3: Cantidad de pacientes según la duración de la estadía')
ax.set_ylabel('Cantidad de Pacientes')
ax.set_xlabel('Duración de la Estadía (en días)')
```

###### **Interpretación:**
Observando el **Gráfico 3**, se puede establecer que la cantidad de días que permanecen en la clínica están entre 1 y 6 días donde el número de pacientes supera las 10 mil personas. Además, en las tres primeras posiciones se encuentra que, en orden descendente, los pacientes permanecen en su gran mayoria 1, 3 o 4 días en la clínica.

##### **Gráfico 4: Boxplot de la cantidad de pacientes según la duración de la estadía**
Para profundizar en el análisis de la duración de la estancia de los pacientes en la clínica, se presenta un boxplot en el cual se podrán visualizar la cantidad de pacientes promedio, los extremos y los outliers que se dan en los tres días que presentan mayor cantidad de pacientes.

```python
x = CANTIDAD_PACIENTES_ESTADIA.loc[CANTIDAD_PACIENTES_ESTADIA['Cantidad_Pacientes'] >= 14000]
fig, ax = plt.subplots(figsize=(8, 4))
ax.boxplot(x=x.T)
plt.xticks([1,2,3], [1, 3, 4])
ax.set_title('Gráfico 4: Boxplot de la cantidad de pacientes según la duración de la estadía')
ax.set_xlabel('Duración de la Estadía (en días)')
ax.set_ylabel('Cantidad de Pacientes')
```

###### **Interpretación:**
Este gráfico boxplot complementa al anterior, ya que se puede observar los valores mínimos y máximos
de la cantidad de pacientes en función a la duración de su estadía en la clínica. También permite  apreciar la media, la mediana y los cuartiles. 
Además, a partir del mismo se puede determinar que esta variable (duración de la estancia) no presentan valores atípicos en estos casos donde se presentan una mayor cantidad de pacientes, es decir, con una duración de la internación de 1, 3 y 4 días.
